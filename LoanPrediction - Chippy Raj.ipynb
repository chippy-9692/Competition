```python
le=LabelEncoder()
loan_train['Gender']=le.fit_transform(loan_train['Gender'])
loan_train['Married']=le.fit_transform(loan_train['Married'])
loan_train['Dependents']=le.fit_transform(loan_train['Dependents'])
loan_train['Self_Employed']=le.fit_transform(loan_train['Self_Employed'])
loan_train['LoanAmount']=le.fit_transform(loan_train['LoanAmount'])
loan_train['Loan_Amount_Term']=le.fit_transform(loan_train['Loan_Amount_Term'])
loan_train['Credit_History']=le.fit_transform(loan_train['Credit_History'])
loan_train['Education']=le.fit_transform(loan_train['Education'])
loan_train['Property_Area']=le.fit_transform(loan_train['Property_Area'])
loan_train['Loan_Status']=le.fit_transform(loan_train['Property_Area'])
```

```python
loan_train.loc[:, 'Gender'].value_counts(dropna=False)
```

```python
loan_train['Gender'].fillna(value='Male', inplace=True)
```

```python
loan_train.drop(['Loan_ID'],axis=1)
```

```python
feature_columns = ['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area','Loan_Status']
```

```python
X = loan_train[feature_columns]
```

```python
y =  loan_train['Loan_Status']
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.01,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf_clf=RandomForestClassifier(n_estimators=50)
rf_clf.fit(X_train,y_train)
```

```python
y_pred_rf=rf_clf.predict(X_test)

#Evaluating the Random Forest model
print('Accuracy=',accuracy_score(y_test,y_pred_rf))
```
